```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('winequality-red.csv')

display(df)

df.info()
```

```python
import matplotlib.pyplot as plt
import numpy as np

colunas_numericas = df.select_dtypes(include=[np.number]).columns

for coluna in colunas_numericas:
    dados_histograma = df[coluna].dropna()  # Remove valores ausentes (NaN)

    # Criando o histograma
    plt.figure(figsize=(8,6))
    plt.hist(dados_histograma, bins=30, edgecolor='black', alpha=0.7, label='cdz')
    
    # TÃ­tulo e eixos personalizados com o nome da coluna
    plt.title(f'Histograma da Coluna: {coluna}')
    plt.xlabel(f'{coluna}')
    plt.ylabel('FrequÃªncia')
    
    #plt.legend()
    plt.show()
```

```python
colunas_numericas = df.select_dtypes(include=[np.number]).columns

# Criando boxplots para cada coluna numÃ©rica
for coluna in colunas_numericas:
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=df[coluna], color='skyblue')
    plt.title(f'Boxplot da Coluna: {coluna}')
    plt.ylabel(coluna)
    plt.grid(True)
    plt.show()
```

```python
# Calcula a matriz de correlaÃ§Ã£o
correlation_matrix = df.corr()

# Cria uma mÃ¡scara para a parte superior
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Configura o tamanho da figura
plt.figure(figsize=(16, 14))

# Cria o heatmap com a mÃ¡scara
sns.heatmap(correlation_matrix, mask=mask, cmap='coolwarm', annot=True, fmt=".2f", linewidths=.5)

plt.title("Matriz de CorrelaÃ§Ã£o (sem duplicatas)")
plt.show()
```
